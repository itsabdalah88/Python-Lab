```python
def add(a,b):
    return a+b

def sub(a,b):
   return a-b

def mul(a,b):
    return a*b

def div(a,b):
    if b !=0:
        return a/b
    else:
        print ("Not Pssible")
        return

def calc(x):
    s = x.split(" ")
    if s[1] == "+":
        print (add(int(s[0]), int(s[-1])))
    elif s[1] == "-":
        print (sub(int(s[0]), int(s[-1])))
    elif s[1] == "*":
        print (mul(int(s[0]), int(s[-1])))
    elif s[1] == "/":
        print (div(int(s[0]), int(s[-1])))    


flag = True 
while flag:
    x = input()
    if x != "done":
        calc(x)
    else:
        flag = False
```

```python
def fact(x):
    if x == 0:
        return 1
    return x*fact(x-1)

def fep(x):
    if x == 1 or x == 2:
        return 1
    else:
        return fep(x-1) + fep(x-2)

def abs(x):
    if x < 0:
        return -x
    else:
        return x


def is_vowel(x):
    count = 0
    vowel = ['a','e','o','u','i', 'A', 'E', 'O', 'U', 'I']
    for i in x:
        if i in vowel:
            count +=1
    return count

def word_count(l):
    count = 0
    for i in l.split():
        count +=1
    return count

def is_pall(w):
    if len(w) == 0 or len(w) == 1:
        return True
    elif w[0] != w[-1]:
        return False
    else:
        return is_pall(w[1:-1])
```
