```python
def calc():
    nums = []
    counter = 0
    while True:
         x = int(input("Enter a number: "))
         if x == 0:
           break
         nums.append(x)
         counter += 1
    print ("The summation of the numbers is: ", sum(nums))
    print ("The average of the numbers: ", sum(nums)/counter)
    print ("The largest number is: ", max(nums))
    print ("The smallest number is: ", min(nums))

calc()

```

```python
def is_pall(w):
    w = w.lower().replace(" ", "")
    
    if len(w) == 0 or len(w) == 1:
        return True
    elif w[0] != w[-1]:
        return False
    else:
        return is_pall(w[1:-1])

is_pall("H iih")
```

```python
def count_char(line, char):
    counter = 0
    for i in line:
        counter += i == char
    return counter

print(count_char("hihihi", "i"))
```

```python
def prime(p):
    if p < 2:
        return False
    for i in range(2, p):
        if p % i == 0:
            return False
    return True

prime(9)
```

```python
def rotate_word(word, rot):
    rotated_char = []
    for i in word:
        if 65 <= ord(i) <=90 :
            n = (((ord(i)-65)+rot)%26) + 65
        elif 97 <= ord(i) <=122 :
            n = (((ord(i)-97)+rot)%26) + 97
        else:
            n = ord(i)
        rotated_char.append(chr(n))
        rotated_word = ''.join(rotated_char)
    print(f'\'{word}\' rotated by {rot} is \'{rotated_word}\'')

print(rotate_word("hello",10))
```
