```python
students = {
    'Ahmed': [85, 90, 78],
    'Sara': [92, 99, 95],
    'Omar': [70, 65, 80],
    'Lina': [60, 75, 68]
}

def cal_avg(st_list):
    return (sum(st_list) / len(st_list))

def grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 80:
        return 'B'
    elif marks >= 70:
        return 'C'
    elif marks >= 60:
        return 'D'
    else:
        return 'F'

def display():
    print("\nAll students: ")
    for name, marks in students.items():
        avg = cal_avg(marks)
        gradee = grade(avg)
        print(f'Name: {name}, Marks: {marks}, Average: {avg}, Grade: {gradee}')

def add_st():
    new_st = input("Name: ")
    marks = []
    for i in range(3):
        m = int(input(f"Enter {i-1} marks"))
        marks.append(m)
        students[new_st] = marks


add_st()
display()
```

```python
import PySimpleGUI as sg
layout = [
    [sg.Text("Enter the first number: "), sg.Input(key='-NUM1-')],
    [sg.Text("Enter the second number: "), sg.Input(key='-NUM2-')],
    [sg.Button("Calculate Sum"), sg.Button("Exit")],
    [sg.Text("", key='-Result-', size=(30,1))]
]

window = sg.Window("Sum Calculator", layout)

while True:
    event, values = window.read()
    if event in (sg.WIN_CLOSED, 'Exit'):
        break

    if event == "Calculate Sum":
        try:
            num1 = float(values['-NUM1-'])
            num2 = float(values['-NUM2-'])
            result = num1 + num2
            window['-Result-'].update(f"Sum: {result}")
        except ValueError:
            window['-Result-'].update("Error: plase enter valid number.")
window.close()
```
